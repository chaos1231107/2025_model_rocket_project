```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
file_path = '/content/drive/MyDrive/가천대/동아리 활동/files/tms1.xlsx'
df = pd.read_excel(file_path)
df
```

```python
plt.plot(df['t'], df['w'], marker='o', linestyle='-')
plt.xlabel('Time(s)')
plt.ylabel('Thrust(N)')
plt.title('Thrust-Time', fontsize=13, y=1.02)
plt.grid(True)
plt.xlim(54.0, 56.0)
plt.show()
```

```python
time = df['t'].values
thrust = df['w'].values

# 54s ~ 56s
mask = (time >= 54) & (time <= 56)
time_filtered = time[mask]
thrust_filtered = thrust[mask]

# 면적 계산
impulse = np.trapezoid(thrust_filtered, time_filtered)

print(f'Total Impulse : {impulse:.2f} N·s')
```
