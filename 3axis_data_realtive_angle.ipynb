```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('MPU_relative.txt', header=None)
df.columns = ['time', 'gyro_roll', 'gyro_pitch', 'gyro_yaw', 'kal_roll', 'kal_pitch', 'kal_yaw',
              'lpf_roll', 'lpf_pitch', 'lpf_yaw']
df
```

```python
#Roll Data
plt.figure(figsize=(20,3))
plt.subplot(1,3,1)
plt.plot(df['time'], df['gyro_roll'], label = 'Raw Data of Roll Axis')
plt.plot(df['time'], df['kal_roll'], label = 'Kalman Filter Roll Axis Data', linestyle='--',linewidth=2.5)
plt.plot(df['time'], df['lpf_roll'], label = 'LPF + KMF Data')
plt.legend()
plt.grid(True)
#Pitch Data
plt.subplot(1,3,2)
plt.plot(df['time'], df['gyro_roll'], label = 'Pitch Data of Roll Axis')
plt.plot(df['time'], df['kal_pitch'], label = 'Kalman Filter Roll Axis Data', linestyle='--', linewidth=2.5)
plt.plot(df['time'], df['lpf_pitch'], label = 'LPF + KMF Data')
plt.legend()
plt.grid(True)
#Yaw Data
plt.subplot(1,3,3)
plt.plot(df['time'], df['gyro_yaw'], label = 'Yaw Data of Roll Axis')
plt.plot(df['time'], df['kal_yaw'], label = 'Kalman Filter Roll Axis Data', linestyle='--', linewidth=2.5)
plt.plot(df['time'], df['lpf_yaw'], label = 'LPF + KMF Data')
plt.legend()
plt.grid(True)
```
