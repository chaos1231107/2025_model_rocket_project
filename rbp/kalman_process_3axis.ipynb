```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('MPU.txt', header=None)
df.columns = ['time', 'gyro_x', 'gyro_y', 'kal_roll', 'kal_pitch']
df
```

```python
#Roll Data
plt.figure(figsize=(10,2))
plt.subplot(1,2,1)
plt.plot(df['time'], df['gyro_x'], label = 'Raw Roll Angle')
plt.plot(df['time'], df['kal_roll'], color= 'red', label = 'Kalman Roll Angle')
plt.title('Roll Axis Angle Data')
plt.grid(True)
plt.legend()

#Pitch Data
plt.subplot(1,2,2)
plt.plot(df['time'], df['gyro_y'], label = 'Raw Pitch Angle')
plt.plot(df['time'], df['kal_pitch'], color='red', label = 'Kalman Pitch Angle')
plt.title('Pitch Axis Angle Data')
plt.grid(True)
plt.legend()
plt.show()
```

```python
df = pd.read_csv('MPU2.txt', header=None)
df
```

```python
df.columns = ['time', 'gyro_x', 'gyro_y','gyro_z', 'kal_roll', 'kal_pitch', 'kal_yaw']
df
```

```python
#Roll Data
plt.figure(figsize=(20,3))
plt.subplot(1,3,1)
plt.plot(df['time'], df['gyro_x'], label = 'Raw Roll Angle')
plt.plot(df['time'], df['kal_roll'], color= 'red', label = 'Kalman Roll Angle')
plt.title('Roll Axis Angle Data')
plt.grid(True)
plt.legend()

#Pitch Data
plt.subplot(1,3,2)
plt.plot(df['time'], df['gyro_y'], label = 'Raw Pitch Angle')
plt.plot(df['time'], df['kal_pitch'], color='red', label = 'Kalman Pitch Angle')
plt.title('Pitch Axis Angle Data')
plt.grid(True)
plt.legend()

#Yaw Data
plt.subplot(1,3,3)
plt.plot(df['time'], df['gyro_z'], label = 'Yaw Pitch Angle')
plt.plot(df['time'], df['kal_yaw'], color='red', label = 'Kalman Yaw Angle')
plt.title('Yaw Axis Angle Data')
plt.grid(True)
plt.legend()
plt.show()
```
