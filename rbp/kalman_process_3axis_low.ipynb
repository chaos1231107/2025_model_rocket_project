```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('MPU.txt', header=None)
df.columns = ['time', 'gyro_x', 'gyro_y', 'kal_roll', 'kal_pitch']
df
```

```python
#Roll Data
plt.figure(figsize=(10,2))
plt.subplot(1,2,1)
plt.plot(df['time'], df['gyro_x'], label = 'Raw Roll Angle')
plt.plot(df['time'], df['kal_roll'], color= 'red', label = 'Kalman Roll Angle')
plt.title('Roll Axis Angle Data')
plt.grid(True)
plt.legend()

#Pitch Data
plt.subplot(1,2,2)
plt.plot(df['time'], df['gyro_y'], label = 'Raw Pitch Angle')
plt.plot(df['time'], df['kal_pitch'], color='red', label = 'Kalman Pitch Angle')
plt.title('Pitch Axis Angle Data')
plt.grid(True)
plt.legend()
plt.show()
```

```python
df = pd.read_csv('MPU2.txt', header=None)
df
```

```python
df.columns = ['time', 'gyro_x', 'gyro_y','gyro_z', 'kal_roll', 'kal_pitch', 'kal_yaw']
df
```

```python
#Roll Data
plt.figure(figsize=(20,3))
plt.subplot(1,3,1)
plt.plot(df['time'], df['gyro_x'], label = 'Raw Roll Angle')
plt.plot(df['time'], df['kal_roll'], color= 'red', label = 'Kalman Roll Angle')
plt.title('Roll Axis Angle Data')
plt.grid(True)
plt.legend()

#Pitch Data
plt.subplot(1,3,2)
plt.plot(df['time'], df['gyro_y'], label = 'Raw Pitch Angle')
plt.plot(df['time'], df['kal_pitch'], color='red', label = 'Kalman Pitch Angle')
plt.title('Pitch Axis Angle Data')
plt.grid(True)
plt.legend()

#Yaw Data
plt.subplot(1,3,3)
plt.plot(df['time'], df['gyro_z'], label = 'Yaw Pitch Angle')
plt.plot(df['time'], df['kal_yaw'], color='red', label = 'Kalman Yaw Angle')
plt.title('Yaw Axis Angle Data')
plt.grid(True)
plt.legend()
plt.show()
```

```python
df2 = pd.read_csv('MPU_lpf2.txt', header=None)
df2
```

```python
df2.columns = ['time', 'roll_raw', 'pitch_raw', 'yaw_raw', 'roll_kal', 'pitch_kal', 'yaw_kal',
               'roll_lpf', 'pitch_lpf', 'yaw_lpf']
df2
```

```python
#Roll Data
plt.figure(figsize=(40,3))
plt.subplot(1,3,1)
plt.plot(df2['time'], df2['roll_raw'], label = 'Raw Roll Angle')
plt.plot(df2['time'], df2['roll_kal'], color= 'red', label = 'Kalman Roll Angle', linewidth=2.5, linestyle='--')
plt.plot(df2['time'], df2['roll_lpf'], color = 'green', label = 'Kalman Filter + Low Pass Filter')
plt.title('Roll Axis Angle Data')
plt.grid(True)
plt.legend()

#Pitch Data
plt.figure(figsize=(40,3))
plt.subplot(1,3,2)
plt.plot(df2['time'], df2['pitch_raw'], label = 'Pitch Roll Angle')
plt.plot(df2['time'], df2['pitch_kal'], color= 'red', label = 'Kalman Pitch Angle', linewidth=2.5, linestyle='--')
plt.plot(df2['time'], df2['pitch_lpf'], color = 'green', label = 'Kalman Filter + Low Pass Filter')
plt.title('Pitch Axis Angle Data')
plt.grid(True)
plt.legend()

#Yaw Data
plt.figure(figsize=(40,3))
plt.subplot(1,3,3)
plt.plot(df2['time'], df2['yaw_raw'], label = 'Pitch Roll Angle')
plt.plot(df2['time'], df2['yaw_kal'], color= 'red', label = 'Kalman Pitch Angle', linewidth=2.5, linestyle='--')
plt.plot(df2['time'], df2['yaw_lpf'], color = 'green', label = 'Kalman Filter + Low Pass Filter')
plt.title('Yaw Axis Angle Data')
plt.grid(True)
plt.legend()
plt.show()

```
